```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
```

```python
df = pd.read_csv("Cars Company.csv")
df
```

## 2. Check and Drop missing or null values

```python
print(df.isnull().sum())
```

```python
df = df.dropna().copy()
df.count()
```

```python
print(df.isnull().sum()) 
```

## 3. What are the top 10 average MSRP (Price) per car brand (rounded figure)? with Horizontal Bar Plot

```python
top_10_msrp = df.groupby("Make")["MSRP"].mean().round().sort_values(ascending=False).head(10)
top_10_msrp
```

```python
top_10_msrp.sort_values().plot(kind='barh', figsize=(10, 5))
plt.title("Top 10 Brands by Average MSRP")
plt.xlabel("Average MSRP")
plt.ylabel("Car Brand")
plt.show()
```

## 4. How has the MSRP (price) changed over the years (rounded figure)? with Line Plot

```python
msrp_over_years = df.groupby("Year")["MSRP"].mean().round().sort_index()
msrp_over_years
```

```python
msrp_over_years.plot(kind='line', figsize=(10, 5))
plt.title('Avg MSRP over years')
plt.xlabel("years")
plt.ylabel("avg MSRP")
plt.show()
```

## 5. How Engine HP (power) and MPG (efficiency) have changed over the years? show with two line plots

Hint: Average of (city mpg + highway MPG)

```python
df["Average_MPG"] = (df["city mpg"] + df["highway MPG"]) / 2
hp_trend = df.groupby("Year")["Engine HP"].mean().round()
mpg_trend = df.groupby("Year")["Average_MPG"].mean().round()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(hp_trend, label='Engine HP')
plt.plot(mpg_trend, label='Average MPG')
plt.title("Engine HP and MPG Trends Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.legend()
plt.show()
```

## 6. What percentage of cars are Compact, Midsize, or Large? Show with Pie chart

```python
size_counts = df["Vehicle Size"].value_counts()
selected_sizes = size_counts[["Compact", "Midsize", "Large"]]
percentages = (selected_sizes / size_counts.sum() * 100).round(2)
```

```python
percentages.plot(kind='pie')
plt.title('Vehicle Size Distribution')
plt.show()
```

## 7. Which vehicle styles are sold the most? show with vertical bar plot

```python
style_counts = df["Vehicle Style"].value_counts()
print(style_counts)

style_counts.plot(kind='bar', figsize=(12, 5), title='Most Sold Vehicle Styles')
plt.xlabel("Vehicle Style")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
filtered_df = df[(df["MSRP"] >= 40000) & (df["Popularity"] >= 1000)]
brand_counts = filtered_df["Make"].value_counts()
brand_counts
```
